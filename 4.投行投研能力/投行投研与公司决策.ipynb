Example:(Expansion Project):

William, Inc. Would like to set up a new plant for Fintech Education.

1.Currently, William has an option to buy an existing building for E-learning at a cost of 24,000.

2.Necessary equipment for the plant will cost 16,000,including installation costs.

3.The equipment falls into a MACRS 5-year class.The building falls into a MACRS 39-year class.

4.The project would require an initial investment of 12,000 in net working capital. The initial working capital investment will be made at the time of the purchases of the building and equipment.

5.The project’s estimated economic life is 4 years. At the end of that time, the building is expected to have a market value of 15,000 and a book value of 21,816, where as the equipment is expected have a market value of 4,000 and a book value of 2,720.

6.Annual sales will be 80,000. The production department has estimated that variable manufacturing costs will total 60% of sales and that fixed overhead costs, excluding depreciation, will be 10,000 a year (costs:0.6*80,000+10,000=58,000). 

7.Depreciation expense will be determined for the year in accordance with the MACRS rate.

8.William’s marginal federal-plus-state tax rate is 40%; its cost of capital is 12%; and, for capital budgeting purposes, the company’s policy is to assume that CFO at the end of each year. The plant will begin operations immediately after the investment is made, and the first CFO will occur exactly one year later.

9.Under MARCRS, the pre-tax depreciation for the building and equipment is：
Year 1=3,512; year 2=5,744; year 3=3,664; year4=2,544

Compute the initial investment outlay, CFO over the project’s life, and the terminal-year cash flows for William’s expansion project. Then determine whether the project should be accepted using NPV analysis:

```python
FCInv=24000+16000
NWCInv=12000

S=80000
C=58000
D_1=3512
D_2=5744
D_3=3664
D_4=2544
T=0.1

Sal_T=15000+4000
B_T=21816+2720

Rate=0.12


def Calculate_Initial_Stage_Out_Flow(FCInv,NWCInv):
    Initial_Stage_Out_Flow=FCInv+NWCInv
    return -Initial_Stage_Out_Flow

def Calculate_CFO(S,C,T,D):
    CFO=(S-C)*(1-T)+T*D
    return CFO

def Calculate_TNOCF(NWCInv,Sal_T,T,B_T):
    TNOCF=Sal_T-T*(Sal_T-B_T)+NWCInv
    return TNOCF

def Calculate_Terminating_Stage_In_Flow(OCF_T,TNOCF):
    Terminating_Stage_In_Flow=OCF_T+TNOCF
    return Terminating_Stage_In_Flow


Calculate_Initial_Stage_Out_Flow(FCInv,NWCInv)

Calculate_CFO(S,C,T,D_1)
Calculate_CFO(S,C,T,D_2)
Calculate_CFO(S,C,T,D_3)
Calculate_CFO(S,C,T,D_4)

Calculate_TNOCF(NWCInv,Sal_T,T,B_T)


CF_LIST=[Calculate_Initial_Stage_Out_Flow(FCInv,NWCInv),
         Calculate_CFO(S,C,T,D_1),Calculate_CFO(S,C,T,D_2)
        ,Calculate_CFO(S,C,T,D_3),
        Calculate_Terminating_Stage_In_Flow(Calculate_CFO(S,C,T,D_4)
        ,Calculate_TNOCF(NWCInv,Sal_T,T,B_T))]


np.npv(Rate,CF_LIST)
```

```python
import numpy as np
project_1=[-150,100,80]
project_2=[-155,80,70,50]

np.npv(0.09,project_1)
np.npv(0.09,project_2)


def calculate_LCM(x,y):
    if x>y:
        num=x
    else:
        num=y
    while(True):
        if ((num % x == 0) and (num % y == 0)):
            LCM = num
            break
        num+=1
        
    return LCM

lcm=calculate_LCM(22,37)

new_project_1=[-150,100,-70,100,-70,100,80]
new_project_2=[-155,80,70,-105,80,70,50]

np.npv(0.000001,new_project_2)
print(lcm)
```
